```python
import cv2
from ipywidgets import Image
from IPython.display import display
```

```python
cap=cv2.VideoCapture("blooms-113004.mp4")
# cap=cv2.VideoCapture(0)
```

```python
width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
fps = cap.get(cv2.CAP_PROP_FPS)
frame_count = cap.get(cv2.CAP_PROP_FRAME_COUNT)

print(f"width:{width} height: {height}")
print(f"fps:{fps}")
print(f"frame count: {frame_count}")
```

```python
video=Image()
display(video)
```

```python
if not cap.isOpened():
    print("동영상을 정상적으로 읽지 못했습니다.")
else:
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        frame = cv2.resize(frame, (640,480))
        # jpg 로 인코딩 먼저 해줘야 한다함
        frame = cv2.imencode(".jpg", frame)[1].tobytes()

        video.value = frame

cap.release()
```

```python
import os

cap=cv2.VideoCapture("blooms-113004.mp4")
```

```python
limit_fps = 10
```

```python
if not cap.isOpened():
    print("동영상을 정상적으로 읽지 못했습니다.")
else:
    save_dir = "./frame_img_save"
    os.makedirs(save_dir, exist_ok=True)
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        now_frame=int(cap.get(1))
        if now_frame % limit_fps == 0:
            cv2.imwrite(f"{save_dir}/img_{str(now_frame).zfill(4)}.png", frame)

        print(f"현재 프레임:{now_frame} / {frame_count}")
cap.release()
```
